```python
import glob, cv2
from glob import glob
from PIL import Image
```

```python
# Directories for our data
train_data_dir = "training_set"
test_data_dir = "testing_set"
```

```python
# Training Data
ellie = glob(train_data_dir+"/ellie/*.jpg")
salem = glob(train_data_dir+"/salem/*.jpg")
ash = glob(train_data_dir+"/ash/*.jpeg")
cleo = glob(train_data_dir+"/cleo/*.jpeg")

# Testing Data
test_e = glob(test_data_dir+"/ellie/*.jpg")
test_s = glob(test_data_dir+"/salem/*.jpg")
test_a = glob(test_data_dir+"/ash/*.jpeg")
test_c = glob(test_data_dir+"/cleo/*.jpeg")

cat_names = ellie + salem + ash + cleo
test_names = test_e + test_s + test_a + test_c
```

```python
for ix, img in enumerate(cat_names):
    print(cat_names[ix])
    img = (cv2.imread(cat_names[ix])[:, :, ::-1])
    img = cv2.resize(img, None, fx=0.074, fy=0.074)
    img = Image.fromarray(img)
    img.save(cat_names[ix])
    print(img.size)
```

```python
for ix, img in enumerate(test_names):
    print(test_names[ix])
    img = (cv2.imread(test_names[ix])[:, :, ::-1])
    img = cv2.resize(img, None, fx=0.074, fy=0.074)
    img = Image.fromarray(img)
    img.save(test_names[ix])
    print(img.size)
```
